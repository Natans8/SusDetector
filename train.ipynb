```python
from ultralytics import YOLO
from ultralytics.data.split import autosplit
from pathlib import Path
import torch
```

```python
torch.cuda.empty_cache()
model = YOLO("models/yolo26n.pt")
```

```python
# Prepare data
data_path = Path("data/run/images")
autosplit(path=data_path, weights=(0.80, 0.10, 0.10), annotated_only=True)
```

```python
out_dir = Path("models/run")
baseline = model.train(
    data='data/run/data.yaml',
    device='cuda',
    epochs=70, # Plateau at about 60-70 epochs
    imgsz=640,
    batch=0.8, # 80% of available VRAM
    save=True,
    save_period=5,
    single_cls=True,
    plots=True,
    project=str(out_dir),
    name='run'
)
```

```python
tuned = model.tune(
    data='data/run/data.yaml',
    device='cuda',
    epochs=60,
    imgsz=640,
    batch=0.8,
    save=True,
    save_period=15,
    single_cls=True,
    iterations=32,
    plots=True,
    project='models',
    name='tune'
)
```
